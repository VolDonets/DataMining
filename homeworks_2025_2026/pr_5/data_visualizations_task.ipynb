## Functions to animate

### 1. Taylor Series Expansion of a Sine Wave
* **The Math:** $$f(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n+1)!} x^{2n+1}$$
* **What to animate:** The number of terms ($n$) in the approximation. As $n$ increases frame-by-frame, the polynomial curve will wrap closer and closer to the true sine wave.
* **Why it's useful:** It visualizes how polynomials can approximate complex functions, which is a core concept in calculus and machine learning optimization.

### 2. Fourier Series Square Wave Synthesis
* **The Math:** $$f(x) = \sum_{k=1,3,5...}^{\infty} \frac{1}{k} \sin(kx)$$
* **What to animate:** The number of odd harmonics ($k$) added to the series. The animation shows smooth sine waves stacking up to eventually form a sharp, blocky square wave.
* **Why it's useful:** It demonstrates signal decomposition. Understanding how complex, noisy data can be broken down into simple periodic frequencies is crucial for time-series analysis.

### 3. Damped Harmonic Oscillator
* **The Math:** $$y(t) = A e^{-\lambda t} \cos(\omega t)$$
* **What to animate:** The damping coefficient $\lambda$. Start with $\lambda = 0$ (endless oscillation), and gradually increase it to show the wave flattening out faster (under-damped to over-damped).
* **Why it's useful:** It introduces exponentially decaying envelopes, which are very common in physics, acoustics, and financial time-series modeling.

### 4. Unit Circle to Sine Tracing
* **The Math:** $$x^2 + y^2 = 1 \rightarrow y = \sin(\theta), x = \cos(\theta)$$
* **What to animate:** A point moving around the circumference of a circle, while simultaneously drawing the resulting sine wave on a second, parallel plot.
* **Why it's useful:** It beautifully bridges geometry and trigonometry. From a coding perspective, it is excellent practice for managing multi-axis (`subplots`) animations synchronously in Matplotlib.

### 5. Polar Rose Curves (Rhodonea)
* **The Math:** $$r = \cos(k\theta)$$
* **What to animate:** The parameter $k$ transitioning smoothly (e.g., from 1 to 5). As $k$ changes, the "petals" of the rose will morph, multiply, and fold into each other.
* **Why it's useful:** It forces students to step away from standard Cartesian grids and work with polar coordinate systems (`projection='polar'`), expanding their plotting vocabulary.

### 6. 3D Ripple Surface (Wave Equation)
* **The Math:** $$z = \sin\left(\sqrt{x^2 + y^2} - v t\right)$$
* **What to animate:** The time variable $t$. The animation should look like a raindrop hitting a pond, with 3D ripples propagating outward from the center $(0,0)$.
* **Why it's useful:** It introduces `mpl_toolkits.mplot3d`. Visualizing 3D surfaces and topographies is an essential skill for understanding cost functions and gradient descent landscapes.

### 7. Quadratic Bézier Curve Construction
* **The Math:** $$B(t) = (1-t)^2 P_0 + 2(1-t)t P_1 + t^2 P_2$$ for $t \in [0,1]$
* **What to animate:** The interpolation parameter $t$. Draw the three control points ($P_0, P_1, P_2$), the moving interpolation lines between them, and the final curve being traced out over time.
* **Why it's useful:** It teaches linear interpolation and parametric geometry. It is a highly visual way to understand how algorithms smooth out data paths.

### 8. Gaussian Probability Density Function (PDF) Morphing
* **The Math:** $$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$
* **What to animate:** First, animate the standard deviation $\sigma$ expanding and contracting to show the bell curve flattening and spiking. Next, animate the mean $\mu$ shifting the curve left and right.
* **Why it's useful:** It deepens intuition for normal distributions, variance, and data spread, which are the bedrock of statistical data mining.

### 9. Shifting Parabola and Root Finding
* **The Math:** $$y = ax^2 + bx + c$$
* **What to animate:** The y-intercept $c$ continuously shifting up and down. 
* **Why it's useful:** A simple but effective way to connect algebraic equations to visual geometry. It clearly shows how roots (where the line crosses $y=0$) appear, merge into a single point, and disappear into complex numbers. It serves as a great stepping stone for animating root-finding algorithms like Newton-Raphson.

### 10. The Lorenz Attractor (Chaos Theory)
* **The Math:** $$\frac{dx}{dt} = \sigma(y-x)$$
    $$\frac{dy}{dt} = x(\rho-z)-y$$
    $$\frac{dz}{dt} = xy-\beta z$$
* **What to animate:** A point tracing its trajectory through 3D space over time based on the differential equations, ideally leaving a fading "tail" behind it to show the chaotic butterfly shape.
* **Why it's useful:** While the math is a step up, it is incredibly rewarding. It introduces systems of differential equations and chaotic systems, and results in one of the most visually stunning plots in mathematics.

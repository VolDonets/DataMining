# Import dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

# Do animation

```python
# 1. Set up the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)

# 2. Create an empty line object that we will update
line, = ax.plot([], [], 'b-', lw=2)

# Lists to store the data points
x_data, y_data = [], []

# 3. Initialization function: plots the background of each frame
def init():
    line.set_data([], [])
    return line,

# 4. Update function: called sequentially for each frame
def update(frame):
    x_data.append(frame)
    y_data.append(np.sin(frame))
    
    # Update the line with new data
    line.set_data(x_data, y_data)
    return line,

# 5. Create the animation object
# frames: an iterable of values to pass to 'update'
# interval: delay between frames in milliseconds
ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 60),
                    init_func=init, blit=True, interval=50)

# 6. Prevent the static figure from displaying alongside the animation
plt.close(fig)

# 7. Display the animation as an interactive HTML widget
HTML(ani.to_jshtml())
```

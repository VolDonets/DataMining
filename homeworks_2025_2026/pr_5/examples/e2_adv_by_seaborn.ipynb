# 0. Install & improt deps

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
```

# 1. Load & prepare the data

```python
# Load the Iris dataset and convert it into a pandas DataFrame
iris_data = load_iris(as_frame=True)
df = iris_data.frame

# Rename columns for easier access
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species_id']
species_names = iris_data.target_names

# Let's look at the first few rows
df.head()
```

# 2. Doing plots

## 2.1. The Pairplot: The Ultimate Exploratory Tool

```python
# Create a new column with actual species names for better plot labels
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species_name'] = df['species_id'].map(species_map)

# Apply a Seaborn theme. This instantly upgrades the look of all our plots.
sns.set_theme(style="whitegrid", palette="muted")
```

```python
# We drop 'species_id' so it isn't plotted as a numeric variable.
# The 'hue' parameter automatically colors the data points by species.
sns.pairplot(df.drop('species_id', axis=1), hue='species_name', height=2.5)

# Matplotlib functions still work to tweak Seaborn plots!
plt.suptitle('Pairwise Relationships in the Iris Dataset', y=1.02)
plt.show()
```

## 2.2. Boxplots: Visualizing Outliers and Quartiles

```python
plt.figure(figsize=(10, 6))

# A boxplot shows the median, quartiles, and outliers of the data
sns.boxplot(x='species_name', y='sepal_width', data=df, palette='Set2')

# Optional: Add a swarmplot on top to see the actual underlying data points!
sns.swarmplot(x='species_name', y='sepal_width', data=df, color=".25", alpha=0.6)

plt.title('Distribution of Sepal Widths by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')

plt.show()
```

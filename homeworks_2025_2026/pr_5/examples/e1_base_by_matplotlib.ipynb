# 0. Install & improt deps

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# 1. Load & prepare the data

```python
# Load the Iris dataset and convert it into a pandas DataFrame
iris_data = load_iris(as_frame=True)
df = iris_data.frame

# Rename columns for easier access
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species_id']
species_names = iris_data.target_names

# Let's look at the first few rows
df.head()
```

# 2. Doing plots

## 2.1. Histogram: Understanding Distributions

```python
# Create a figure with a specific size (width, height)
plt.figure(figsize=(8, 5))

# Plot the histogram for petal length
plt.hist(df['petal_length'], bins=20, color='skyblue', edgecolor='black')

# Always include title and axis labels!
plt.title('Distribution of Petal Lengths in the Iris Dataset')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency (Number of Flowers)')

# Display the plot
plt.show()
```

## 2.2. Scatter Plot: Identifying Relationships

```python
plt.figure(figsize=(9, 6))

# Define a color for each species
colors = ['red', 'green', 'blue']

# Loop through each species to plot them separately so we can generate a legend
for i, species in enumerate(species_names):
    # Filter the DataFrame for the current species
    subset = df[df['species_id'] == i]
    
    # Create the scatter plot for this subset
    plt.scatter(subset['sepal_length'], subset['sepal_width'], 
                c=colors[i], label=species, alpha=0.7, edgecolors='white', s=70)

# Add descriptions
plt.title('Sepal Length vs. Sepal Width by Iris Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Add a legend using the labels defined in plt.scatter()
plt.legend(title='Species')

# Optional: add a light grid for readability
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

# 2.3. Bar Chart: Comparing Aggregated Data

```python
# First, we need to aggregate our data. 
# We group by the species ID and calculate the mean for all numeric columns.
mean_values = df.groupby('species_id').mean()

plt.figure(figsize=(8, 5))

# Plot a bar chart: X-axis is species names, Y-axis is the calculated mean petal width
plt.bar(species_names, mean_values['petal_width'], 
        color=['#FF9999', '#66B2FF', '#99FF99'], edgecolor='black')

# Add descriptions
plt.title('Average Petal Width per Iris Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Width (cm)')

plt.show()
```

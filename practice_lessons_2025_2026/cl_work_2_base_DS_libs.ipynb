```python
# do data norm (min-max norm)

print(f"Before norm: {arr_1d_2}")

arr_1d_2_norm = (arr_1d_2 - arr_1d_2.min())/ (arr_1d_2.max() - arr_1d_2.min())
print(f"After norm: {arr_1d_2_norm}")
```

```python
# do data norm (standard norm)

print(f"Before norm: {arr_1d_2}")

arr_1d_2_norm = (arr_1d_2 - arr_1d_2.mean()) / np.std(arr_1d_2)
print(f"After norm: {arr_1d_2_norm}")
```

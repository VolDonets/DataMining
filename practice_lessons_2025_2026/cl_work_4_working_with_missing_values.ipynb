```python
df_copy["Name"] = df_copy["Name"].fillna("Unmentioned")
```

```python
df_copy.head(10)
```

```python
df_copy.tail(10)
```

```python
obj_cols = df_copy.select_dtypes(include=['object']).columns
obj_cols
```

```python
df_copy[obj_cols] = df_copy[obj_cols].fillna("MISSED")
df_copy.head(10)
```

```python
# Suppose we have a DataFrame with numeric (or object values only)
# df.fillna(-300, inplace=True)
```

```python
df_copy = df.copy(deep=True)
df_copy.head(10)
```

```python
# 2. Fill 'Female' for missing rows where Name contains "Miss" or "Mrs."
df_copy.loc[df_copy['Sex'].isna() & df_copy['Name'].str.contains('Miss|Mrs\.', regex=True), 'Sex'] = 'FEMALE'

# 3. Fill 'Male' for missing rows where Name contains "Mr."
df_copy.loc[df_copy['Sex'].isna() & df_copy['Name'].str.contains('Mr\.', regex=True), 'Sex'] = 'MALE'
```

```python
df_copy.head(10)
```

```python
df_copy.to_csv("./src/filled_sex_by_name.csv", index=False)
```

# II strategy: removing columns or rows by missing values statistics

```python
df_copy = df.copy(deep=True)
```

```python
missing_percentage = (df_copy.isna().mean() * 100).round(2)
missing_percentage
```

```python
threshold = 50   # in practice, it could be more than 95%
                 # 97-99%
missing_percentage = missing_percentage[missing_percentage > threshold]
missing_percentage
```

```python
list(missing_percentage.index)
```

```python
df_copy.drop(columns=list(missing_percentage.index), inplace=True)
df_copy.head(10)
```

```python
missing_percentage = (df_copy.isna().mean() * 100).round(2)
missing_percentage
```

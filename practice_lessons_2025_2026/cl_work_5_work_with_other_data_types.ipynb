# 2. Working with Date & DateTime

1. International Standard Formats
    
   SO 8601: The global standard for date and time representation.
   
     * `YYYY-MM-DD` (e.g., 2026-03-26).
     * `YYYY-MM-DDTHH:mm:ss.sssZ` (e.g., 2026-03-26T12:29:00.000Z).
        
    RFC 2822 / RFC 1123: Used primarily in HTTP headers and email.
        
     * `ddd, DD MMM YYYY HH:mm:ss GMT` (e.g., Thu, 26 Mar 2026 12:29:00 GMT). 

2. Common Regional Formats
Formats vary significantly by country and culture: 

    US (MDY): `MM/DD/YYYY` (e.g., 03/26/2026).

    Europe/UK (DMY): `DD/MM/YYYY` (e.g., 26/03/2026).
   
    China/Japan/Korea (YMD): `YYYY/MM/DD` (e.g., 2026/03/26). 

3. Custom Format Tokens
Programming languages use specific "tokens" to build custom date strings: 

    Year: `yy` (2-digit), `yyyy` (4-digit).

    Month: `M` (1-digit), `MM` (2-digit), `MMM` (short name), `MMMM`(full name).
   
    Day: `d` (1-digit), `dd` (2-digit), `ddd` (short weekday), `dddd` (full weekday).
   
    Time: `HH` (24-hour), `hh` (12-hour), `mm` (minutes), `ss` (seconds), a or `tt` (AM/PM). 

4. Language-Specific Methods

    JavaScript: Use native methods like `toISOString()` for standards or `toLocaleDateString()` for local formats.
   
    Python: Use `strftime()` to define explicit formats like `"%Y-%m-%d %H:%M:%S"`.
   
    Java: Use the DateTimeFormatter class to parse and format strings.
   
    .NET: Supports Standard Format Strings (e.g., "d" for short date) and Custom Format Strings (e.g., `"yyyy-MM-dd"`).

```python
df["TimeStamp_month"] = df["TimeStamp_datetime"].dt.month
df.head()
```

```python
df["TimeStamp_day"] = df["TimeStamp_datetime"].dt.month
df.head()
```

```python
df["TimeStamp_hour"] = df["TimeStamp_datetime"].dt.hour
df.head()
```

```python
today_ts = pd.Timestamp.today()
today_ts
```

```python
df["Duration_year"] = (today_ts - df["TimeStamp_datetime"].dt.tz_localize(None)).dt.days / 365.25
df.head()
```

```python
df["Assembled_date"] = df['TimeStamp_year'].astype(str) + "_" + df['TimeStamp_day'].astype(str) + "_" + df['TimeStamp_month'].astype(str)
df.head()
```

```python
df["Assembled_datetime"] = pd.to_datetime(df["Assembled_date"], format="%Y_%d_%m")
df.head()
```

# 3. Categorical data (boolean)

Presentations of such data:

* `Y` & `N`
* `T` & `F`
* `True` & `False`
* `A` & `B`

```python
df = pd.DataFrame({
    "BinCol": ["Y", "N", "Y", "Y", "Y", "N"]
})
df.head(6)
```

```python
df["BinCol"].value_counts()
```

```python
df["BinCol_Bool"] = df["BinCol"] == "Y"
df.head()
```

```python
df["BinCol_Int"] = (df["BinCol"] == "Y").astype(int)
df.head()
```

```python
df["BinCol_MapValue"] = df["BinCol"].map(lambda x: 1 if x == "Y" else -1)
df.head()
```

```python
df["BinCol_MapValueDict"] = df["BinCol"].map({"Y": 1, "N": -1})
df.head()
```

```python
df.dtypes
```

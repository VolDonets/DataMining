# Fill missed values with 
1. Special missed value
2. Neighbour UP, BOTTOM
3. Mean
4. Median

## 1. Special missed values

```python
df = iris_df.copy(deep=True)
```

```python
df["sepal_length"] = df["sepal_length"].fillna(value=-1.)
df["sepal_width"] = df["sepal_width"].fillna(value=-1.)
df["petal_length"] = df["petal_length"].fillna(value=-1.)
df["petal_width"] = df["petal_width"].fillna(value=-1.)
df["class"] = df["class"].fillna(value="UNKNOWN")
```

## 3. & 4. Filling with Mean & Median values

```python
df = iris_df.copy(deep=True)
```

```python
iris_df.head(10)
```

```python
a_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
df = df[a_cols].fillna(df[a_cols].mean())
df["class"] = iris_df["class"]
df.head(10)
```

```python
df = iris_df.copy(deep=True)
```

```python
iris_df.head(10)
```

```python
a_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
df = df[a_cols].fillna(df[a_cols].median())
df["class"] = iris_df["class"]
df.head(10)
```
